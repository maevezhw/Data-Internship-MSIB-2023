```python
def search_for(nama_df, kategori):
    data_sem = pd.DataFrame()
    for i, row in enumerate(nama_df):
        ada = False
        for j in row:
            if kategori in j: 
                data_sem = data_sem.append({'Index': i, f'{kategori}': j}, ignore_index= True)
                ada = True
        if (ada == False):
            data_sem = data_sem.append({'Index': i, f'{kategori}': None}, ignore_index= True)

    return data_sem
```

```python
def add_sem (nama_df, df_before, kategori):
    data_sem = search_for(nama_df, kategori)
    data_sem = data_sem.groupby('Index')[kategori].apply(list).reset_index()
    data = pd.merge(df_before, data_sem, left_on=df.index, right_on='Index', how='left')
    data = data.drop(columns = 'Index')

    return data
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from collections import Counter
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Fetch Indonesian stopwords
stop_words = set(stopwords.words('indonesian'))

# Stemming
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Tokenization
df["Kriteria"] = df["Kriteria"].apply(lambda x: [re.sub(r'[^\w\s]', ' ', word) for word in x])
df["tokens"] = df["Kriteria"].apply(lambda x: [word_tokenize(sentence) for sentence in x])

# Removing stopwords
df["tokens"] = df["tokens"].apply(lambda x: [[word for word in sentence if word.lower() not in stop_words] for sentence in x])

# Stemming


df['tokens'] = df['tokens'].apply(lambda x: [[stemmer.stem(word) for word in sentence] for sentence in x])

# Combining all tokens into a single list
all_tokens = [token for sublist in df["tokens"].sum() for token in sublist]

# Counting word occurrences
word_counts = Counter(all_tokens)

list_word = pd.DataFrame()
# Displaying the results
print("Word counts per word:")
for word, count in word_counts.items():
    print(f"{word}: {count}")
    list_word = list_word.append({"Kata": word, "Count": count}, ignore_index = True)
```
