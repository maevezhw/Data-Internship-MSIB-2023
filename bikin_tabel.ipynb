```python
df_magang = pd.read_csv("Informasi Magang 2.csv", index_col = [0])
df_magang.head()
```

```python
daftar_kategori = list(set(df_magang['Kategori']))
daftar_kategori
```

```python
df_magang['Kategori'] = df_magang['Kategori'].replace('Data Analytic', 'Data Analyst')
df_magang['Kategori'] = df_magang['Kategori'].replace('Analis Data', 'Data Analyst')
df_magang['Kategori'] = df_magang['Kategori'].replace('Machine Learning', 'ML Engineer')
df_magang['Kategori'] = df_magang['Kategori'].replace('Artificial Intelligence', 'AI Engineer')
```

```python
daftar_kategori = list(set(df_magang['Kategori']))
daftar_kategori
```

```python
df_kategori = pd.DataFrame()
df_kategori['Kategori'] = daftar_kategori
df_kategori['id_kategori'] = None

for i, row in enumerate(daftar_kategori):
    df_kategori['id_kategori'][i] = f"K-{i+1}"
```

```python
df_kategori = df_kategori[['id_kategori', 'Kategori']]
df_kategori.to_csv('List Kategori.csv', index = False)
```

```python
daftar_perusahaan = list(set(df_magang['Perusahaan']))
print(daftar_perusahaan)

df_perusahaan = pd.DataFrame()
df_perusahaan['Perusahaan'] = daftar_perusahaan
df_perusahaan['id_perusahaan'] = None

for i, row in enumerate(daftar_perusahaan):
    df_perusahaan['id_perusahaan'][i] = f"P-{i+1}"

df_perusahaan = df_perusahaan[['id_perusahaan', 'Perusahaan']]
df_perusahaan.to_csv('List Perusahaan.csv', index = False)
```

```python
df_magang['Semester'] = df_magang['Semester'].apply(lambda x: x[1] if x[1:-1] != 'No Requirement' else x[1:-1])
df_magang['IPK'] = df_magang['IPK'].apply(lambda x: x[1:-1])
df_magang.drop('Kriteria', axis = 1, inplace = True)
df_magang.head(5)
```

```python
df_magang['id_magang'] = None

for i, row in df_magang.iterrows():
    df_magang['id_magang'][i] = f"ID-{i+1}"

df_magang.head()
```

```python
for i, rowpos in df_magang.iterrows():
    for j, rowper in df_perusahaan.iterrows():
        if (rowpos['Perusahaan'] == rowper['Perusahaan']):
            df_magang['Perusahaan'][i] = df_perusahaan['id_perusahaan'][j]

df_magang.head()
```

```python
for i, rowm in df_magang.iterrows():
    for j, rowkat in df_kategori.iterrows():
        if (rowm['Kategori'] == rowkat['Kategori']):
            df_magang['Kategori'][i] = df_kategori['id_kategori'][j]

df_magang
```

```python
df_magang.rename(columns = {'Perusahaan': 'id_perusahaan', 'Kategori' : 'id_kategori'}, inplace= True)
```

```python
df_posisi = df_magang[['id_magang', 'id_kategori', 'Posisi', 'id_perusahaan', 'Lokasi', 'Semester', 'IPK']]
df_posisi.to_csv('List Magang.csv', index = False)
```

```python
df_magang['Jurusan'] = df_magang['Jurusan'].apply(lambda x: eval(x))
df_magang['Soft Skills'] = df_magang['Soft Skills'].apply(lambda x: eval(x))
df_magang['Hard Skills'] = df_magang['Hard Skills'].apply(lambda x: eval(x))
df_magang.head()
```

```python
list_jurusan = []

for i, rowm in df_magang.iterrows():
    for j in rowm['Jurusan']:
        list_jurusan.append({'id_magang': rowm['id_magang'], 'Jurusan': j})

list_jurusan = pd.DataFrame(list_jurusan)
list_jurusan
```

```python
list_jurusan.drop(list_jurusan[list_jurusan['Jurusan']==""].index, inplace=True)
```

```python
list_jurusan.to_csv('Jurusan per Posisi.csv', index = False)
```
