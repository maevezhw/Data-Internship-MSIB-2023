```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.expected_conditions import presence_of_element_located
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support import expected_conditions as EC
import warnings
import time 
import string
# import openpyxl
import os

warnings.filterwarnings("ignore")
```

```python
import pandas as pd
data_df = pd.DataFrame()
it = 0
```

```python
from selenium.common.exceptions import StaleElementReferenceException, ElementNotInteractableException
#Loading Selenium Webdriver 
driver= webdriver.Chrome()
wait = WebDriverWait(driver, 5)
 
#Opening Google maps 
driver.get("https://kampusmerdeka.kemdikbud.go.id/program/magang/browse/")
time.sleep(3)

searchbox = driver.find_element(By.CSS_SELECTOR, "input[placeholder='Semua Posisi']")

category = ["Data Analyst", "Data Scientist", "ML Engineer", "Data Engineer", "AI Engineer", "Analis Data", "Data Analytic", "Machine Learning", "Artificial Intelligence"]

for i, cat in enumerate(category): 
    if (i >= it):
        searchbox.send_keys(cat)
        searchbox.send_keys(Keys.ENTER)
        time.sleep(5)

        sidebar = driver.find_element(By.CSS_SELECTOR, ".box-0-0-1.flex-0-0-2.container-0-0-180")

        # Locate the sidebar element
        # sidebar = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "box-0-0-1 content-0-0-182")))

        # Get the initial number of elements in the sidebar
        initial_element_count = len(sidebar.find_elements(By.CSS_SELECTOR, ".box-0-0-1.row-0-0-4.clickable-0-0-8.transition-0-0-76"))
        #print(initial_element_count)
        # Simulate scrolling until no more elements are loaded
        while True:
            driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight;", sidebar)

            time.sleep(5)
            # WebDriverWait(driver, 10)
            
            # Get the updated number of elements in the sidebar
            updated_element_count = len(sidebar.find_elements(By.CSS_SELECTOR, ".box-0-0-1.row-0-0-4.clickable-0-0-8.transition-0-0-76"))
            # print(initial_element_count, updated_element_count)

            if updated_element_count == initial_element_count:
                # print("break:", updated_element_count)
                break  # No more elements are being loaded, exit the loop
            else:
                initial_element_count = updated_element_count  # Update the initial count for the next iteration
        
        entries = driver.find_elements(By.CSS_SELECTOR, ".box-0-0-1.row-0-0-4.clickable-0-0-8.transition-0-0-76")

        for entry in entries: 
            identitas = entry.find_elements(By.CSS_SELECTOR, ".text-0-0-26.sans-0-0-40")
            posisi = identitas[0].text
            lokasi = identitas[3].text
            print(posisi, lokasi)

            entry.click()
            time.sleep(3)
            
            for window_handle in driver.window_handles:
                if window_handle != driver.current_window_handle:
                    driver.switch_to.window(window_handle)
                    break

            detail = driver.find_elements(By.CSS_SELECTOR, ".box-0-0-1.content-0-0-182.content-0-0-236")
            krit = detail[0].find_elements(By.CSS_SELECTOR, ".text-0-0-26.sans-0-0-40.small-0-0-43.justify-0-0-34.pt4-0-0-751")
            kriteria = krit[1].text

            data_df = data_df.append({'Kategori': cat, 'Posisi': posisi, 'Perusahaan': lokasi.split("•")[0].strip(), 'Lokasi': lokasi.split("•")[1].strip(), 'Kriteria': kriteria}, ignore_index=True)

        time.sleep(5)
        it += 1
        searchbox.send_keys(Keys.CONTROL + 'a')  # Select all text
        searchbox.send_keys(Keys.BACKSPACE) 
        WebDriverWait(driver, 10)


data_df
```

```python
data_df = data_df.drop_duplicates().reset_index(drop=True)
data_df
```
